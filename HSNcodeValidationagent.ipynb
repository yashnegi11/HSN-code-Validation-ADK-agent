```python
#Loading Excel data into a Pandas DataFrame
import pandas as pd
import io

hsn_df = pd.read_excel("HSN_SAC.xlsx", dtype=str)
hsn_df.dropna(subset=['HSNCode'], inplace=True)
hsn_dict = dict(zip(hsn_df['HSNCode'], hsn_df['Description']))
hsn_df.head()
```

```python
# Logic
def is_numeric(code):
    return code.isdigit() and len(code) in [2, 4, 6, 8]

def exist_in_master(code):
    return code in hsn_dict

def check_hierarchy(code):
    parents = [code[:i] for i in [2, 4, 6] if i < len(code)]
    return all(p in hsn_dict for p in parents)
```

```python
#Agent
def validate_hsn_codes(codes):
    for code in codes:
        if not is_numeric(code):
            print(f"HSN code {code} is invalid: Incorrect format.")
        elif not exist_in_master(code):
            print(f"HSN code {code} is invalid: Not found in master data.")
        else:
            desc = hsn_dict[code]
            print(f"HSN code  {code} is valid: {desc}")
            if len(code) == 8:
                if check_hierarchy(code):
                    print(" AND All PARENT CODES EXIST.")
                else:
                    print(" Some parent codes are missing.")
```

```python
# Test
test_codes = ['0101', '01012100', '999999']
validate_hsn_codes(test_codes)
```
